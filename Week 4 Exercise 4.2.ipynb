```python
class SimpleDB:
    def __init__(self):
        self.db = dict()

    def insert(self, key, value):
        if key not in self.db:
            self.db[key] = value
            return True
        else:
            self.db[key] = value
            return False

    def delete(self, key):
        if key in self.db:
            del self.db[key]
            return True
        else:
            return False

    def fetch(self, key):
        return self.db.get(key, None)

# Populate the `people` list with the data provided 
people = [
    {'username': 'ryanandrew', 'name': 'Lisa Weber', 'age': 64, 'follows': ['jill95', 'jeffrey39']},
    {'username': 'robertkirk', 'name': 'Austin Harris', 'age': 21, 'follows': ['achen', 'stricklandheather']},
    {'username': 'jill95', 'name': 'Jason Tran', 'age': 72, 'follows': ['paul31', 'achen', 'uguerreo', 'murphydanny']},
    {'username': 'uguerreo', 'name': 'Jason Marshall', 'age': 45, 'follows': ['ryanandrew', 'achen']},
    {'username': 'pparker', 'name': 'Aaron Elliott', 'age': 21, 'follows': ['paul31']},
    {'username': 'xwilliams', 'name': 'John Dudley', 'age': 12, 'follows': ['ryanandrew', 'foleyangela', 'jeffrey39', 'alisonkeith']},
    {'username': 'kerrjulie', 'name': 'Charles Roberts', 'age': 35, 'follows': ['paul31']},
    {'username': 'stricklandheather', 'name': 'Sherry Nguyen', 'age': 27, 'follows': ['paul31', 'alisonkeith']},
    {'username': 'achen', 'name': 'Dwayne Hanson', 'age': 86, 'follows': ['uguerreo', 'xwilliams']},
    {'username': 'jeffrey39', 'name': 'James Henderson', 'age': 11, 'follows': ['murphydanny']},
    {'username': 'alisonkeith', 'name': 'Jordan Jordan', 'age': 39, 'follows': ['uguerreo']},
    {'username': 'murphydanny', 'name': 'Cindy Brown', 'age': 37, 'follows': ['ryanandrew', 'foleyangela', 'achen']},
    {'username': 'mgiles', 'name': 'Dawn Lopez', 'age': 44, 'follows': ['ryanandrew']},
    {'username': 'paul31', 'name': 'Jesus Thomas', 'age': 18, 'follows': ['robertkirk']},
    {'username': 'foleyangela', 'name': 'Juan Wood', 'age': 59, 'follows': ['achen', 'jeffrey39']}
]

# Create an instance of SimpleDB
simple_db = SimpleDB()

# Insert all profiles into the database
for person in people:
    username = person['username']
    insert_result = simple_db.insert(username, person)
    print(f"Inserted {username}: ", insert_result)

# Fetch and print all profiles to verify insertion
for person in people:
    username = person['username']
    profile = simple_db.fetch(username)
    print(f"Profile for {username}: ", profile)

# Test delete method for the first username
delete_result = simple_db.delete(people[0]['username'])
print(f"Deleted {people[0]['username']}: ", delete_result)

# Fetch the deleted profile to verify
fetch_deleted_profile = simple_db.fetch(people[0]['username'])
print(f"Fetch deleted profile for {people[0]['username']}: ", fetch_deleted_profile)
```

#### Assignment 4.2

```python
from pathlib import Path
from tinydb import TinyDB, Query

# Set up the directory for the database file
output_dir = Path('output')
output_dir.mkdir(parents=True, exist_ok=True)
db_path = output_dir.joinpath('tinydb-people.json')

# Create a TinyDB instance and clear any existing data
people_db = TinyDB(db_path)
people_db.truncate()

# Insert all entries from the `people` dataset into the newly created database
for person in people:
    people_db.insert(person)

# Perform a search that returns all people older than 40
# and assign the results to over_40_results.
Query = Query()
over_40_results = people_db.search(Query.age > 40)

# Showing the over 40 results
print(over_40_results)

# Remove all people older than 40 from the database
# Verify removal by performing an additional search.
people_db.remove(Query.age > 40)
# Verify the removal, which should now be an empty list
verification_results = people_db.search(Query.age > 40)

print(verification_results)
```

The result shows that the query for finding all people older than 40 in the TinyDB database was successful. The search returned a list of dictionaries, each representing a person over 40 years old. Thereafter, the removal operation was executed, and a verification search confirmed that no entries for people older than 40 remained in the database

#### Assignment 4.3

```python
from cog.torque import Graph

# Initialize the graph
g = Graph("people")

# Insert the people data into the graph
for person in people:
    username = person['username']
    for follows in person['follows']:
        g.put(username, "follows", follows)

# Display the relationships between people
g.v().tag("from").out("follows").tag("to").view("follows").render()

# Query to find the usernames of people who follow 'murphydanny'
query_result = g.v("murphydanny").inc("follows").tag("from").all()

# The query result is a dictionary containing a 'result' key,
# which is a list of dictionaries with 'id' and 'from' keys
murphydanny_followers = [person['from'] for person in query_result['result']]
print(murphydanny_followers)
```

The graph database accurately captures the follow relationships among users, with nodes and edges representing the users and their connections, respectively. The visual representation confirms the directional "follows" links from one user to another. Furthermore, the code successfully executed a targeted query to identify users following "murphydanny," yielding "jeffrey39" and "jill95" as the result.
