The task is to create a Directed Acyclic Graph (DAG) using the NetworkX library in Python, which will represent the workflow. The workflow consists of tasks and datasets. In this DAG, nodes represent either tasks or datasets, and edges represent dependencies (where a task must be completed before moving on to the next or generating a dataset).

#### Assignment 6.1a

```python
import networkx as nx

# Create a new directed graph which will represent the workflow
workflow_graph = nx.DiGraph()

# Task nodes
agg_sales_task = 'AggSalesTask'
latest_demo_task = 'LatestDemoTask'
agg_weather_task = 'AggWeatherTask'
modeling_data_task = 'ModelingTask'

# Dataset nodes
sales_data = ['SalesData0', 'SalesData1', 'SalesData2', 'SalesData3', 'SalesData4', 'SalesData5', 'SalesData6']
forecast_data = ['ForecastData0', 'ForecastData1', 'ForecastData2', 'ForecastData3', 'ForecastData4', 'ForecastData5', 'ForecastData6']
agg_sales_data = 'AggSalesData'
agg_weather_data = 'AggWeatherData'
latest_demo_data = 'LatestDemoData'
modeling_data = 'ModelingData'
latest_pums_data = 'LatestPUMSData'
latest_acs_data = 'LatestACSData'
latest_tiger_data = 'LatestTigerData'

# Add dataset nodes to the graph
for dataset in sales_data + forecast_data:
    workflow_graph.add_node(dataset, dtype='data')

workflow_graph.add_node(agg_sales_data, dtype='data')
workflow_graph.add_node(agg_weather_data, dtype='data')
workflow_graph.add_node(latest_demo_data, dtype='data')
workflow_graph.add_node(modeling_data, dtype='data')
workflow_graph.add_node(latest_pums_data, dtype='data')
workflow_graph.add_node(latest_acs_data, dtype='data')
workflow_graph.add_node(latest_tiger_data, dtype='data')

# Add task nodes to the graph
workflow_graph.add_node(agg_sales_task, dtype='task')
workflow_graph.add_node(latest_demo_task, dtype='task')
workflow_graph.add_node(agg_weather_task, dtype='task')
workflow_graph.add_node(modeling_data_task, dtype='task')

# Print the number of nodes and edges in the graph
print(f'Number of nodes: {len(workflow_graph.nodes)}, Number of edges: {len(workflow_graph.edges)}')
```

Added nodes for both tasks and datasets to the “workflow_graph”, and tagged them with an attribute dtype to differentiate between “task” nodes and “data” nodes. Initialized the graph with the required nodes. Output shows 25 nodes with no edges between them yet.

#### Assignment 6.1b

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a new directed graph
workflow_graph = nx.DiGraph()

# Task nodes
agg_sales_task = 'AggSalesTask'
latest_demo_task = 'LatestDemoTask'
agg_weather_task = 'AggWeatherTask'
modeling_data_task = 'ModelingTask'

# Dataset nodes
sales_data = ['SalesData0', 'SalesData1', 'SalesData2', 'SalesData3', 'SalesData4', 'SalesData5', 'SalesData6']
forecast_data = ['ForecastData0', 'ForecastData1', 'ForecastData2', 'ForecastData3', 'ForecastData4', 'ForecastData5', 'ForecastData6']
agg_sales_data = 'AggSalesData'
agg_weather_data = 'AggWeatherData'
latest_demo_data = 'LatestDemoData'
modeling_data = 'ModelingData'
latest_pums_data = 'LatestPUMSData'
latest_acs_data = 'LatestACSData'
latest_tiger_data = 'LatestTigerData'

# Add dataset nodes to the graph
for dataset in sales_data + forecast_data:
    workflow_graph.add_node(dataset, dtype='data')

workflow_graph.add_node(agg_sales_data, dtype='data')
workflow_graph.add_node(agg_weather_data, dtype='data')
workflow_graph.add_node(latest_demo_data, dtype='data')
workflow_graph.add_node(modeling_data, dtype='data')
workflow_graph.add_node(latest_pums_data, dtype='data')
workflow_graph.add_node(latest_acs_data, dtype='data')
workflow_graph.add_node(latest_tiger_data, dtype='data')

# Add task nodes to the graph
workflow_graph.add_node(agg_sales_task, dtype='task')
workflow_graph.add_node(latest_demo_task, dtype='task')
workflow_graph.add_node(agg_weather_task, dtype='task')
workflow_graph.add_node(modeling_data_task, dtype='task')

# Add edges for `agg_sales_task` (inputs and output)
for day_sales in sales_data:
    workflow_graph.add_edge(day_sales, agg_sales_task)
workflow_graph.add_edge(agg_sales_task, agg_sales_data)

# Add edges for `agg_weather_task` (inputs and output)
for day_forecast in forecast_data:
    workflow_graph.add_edge(day_forecast, agg_weather_task)
workflow_graph.add_edge(agg_weather_task, agg_weather_data)

# Add edges for `latest_demo_task` (inputs and output)
workflow_graph.add_edge(latest_pums_data, latest_demo_task)
workflow_graph.add_edge(latest_tiger_data, latest_demo_task)
workflow_graph.add_edge(latest_acs_data, latest_demo_task)
workflow_graph.add_edge(latest_demo_task, latest_demo_data)

# Add edges for `modeling_data_task` (inputs and output)
workflow_graph.add_edge(agg_sales_data, modeling_data_task)
workflow_graph.add_edge(agg_weather_data, modeling_data_task)
workflow_graph.add_edge(latest_demo_data, modeling_data_task)
workflow_graph.add_edge(modeling_data_task, modeling_data)

# Print the number of nodes and edges in the graph
print(f'Number of nodes: {len(workflow_graph.nodes)}, Number of edges: {len(workflow_graph.edges)}')

# Provide graph to visualize the workflow
plt.figure(figsize=(12, 8))
nx.draw(workflow_graph, with_labels=True, node_size=2000, node_color="lightblue", font_size=12, arrowsize=20)
plt.show()
```

Created the DAG with nodes representing each dataset and task, and edges representing the dependencies between them. The visual representation using matplotlib confirms the structure of the DAG by showing 25 nodes, which include both task and dataset nodes, and 24 edges representing the flow of data from datasets to tasks and the generation of new datasets from these tasks.
