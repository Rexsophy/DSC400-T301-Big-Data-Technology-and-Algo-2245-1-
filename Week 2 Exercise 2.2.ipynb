```python
import pandas as pd

# Define the URL of the JSON file containing quiz answers
quiz_answers_json = 'https://raw.githubusercontent.com/pixelastic/pokemonorbigdata/master/app/questions.json'

# Read the JSON file into a Pandas DataFrame
df_all = pd.read_json(quiz_answers_json)

# Filter for Pokémon answers (type 0)
pokemon_answers = df_all[df_all['type'] == 0]

# Filter for Big Data answers (type 1)
big_data_answers = df_all[df_all['type'] == 1]

# Print first few rows of each:
print("Pokemon Answers:")
print(pokemon_answers.head())

print("\nBig Data Answers:")
print(big_data_answers.head())
```

This preliminary task is designed to hone data manipulation skills utilizing Python and the Pandas library, focusing on the handling of JSON data—a format ubiquitous in web APIs and Big Data workflows. The exercise of fetching and categorizing quiz answers serves as a practical application of DataFrame operations and allows one to efficiently filter and organize data. This is probably crucial and preparatory for tackling the complex challenges presented by Big Data analytics.
